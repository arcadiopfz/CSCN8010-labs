# 1. Graph using Matplotlib

![Titanic departing Southampton on 10 April 1912.](https://upload.wikimedia.org/wikipedia/commons/thumb/f/fd/RMS_Titanic_3.jpg/300px-RMS_Titanic_3.jpg)

To create the graph below, we will use the plotting library for Python called Matplotlib. As data, we will use the classic Titanic database, containing the number of passengers, age, sex, survivors, etc.

For more information about the Titanic you can access the following [link](https://en.wikipedia.org/wiki/Titanic).

The graph is a histogram showing the distribution of the number of passengers and their age.

```python
# Importing several libraries for data visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Loading the dataset in seaborn data repository of Titanic
df = sns.load_dataset('titanic')

df.head()
```

```python
# The 'age' column was selected and missing values were dropped by using .dropna()
ages = df['age'].dropna()

#The number of bins were set and to create the histogram
n_bins = 30

# Creating the histogram plot
plt.hist(ages, bins=n_bins, edgecolor="white")


# Setting labels and title
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Histogram of Passenger Ages')

# Showing the plot
plt.show()
```

# 2. Graph using Seaborn

![Titanic at the docks of Southampton.](https://upload.wikimedia.org/wikipedia/commons/thumb/2/2a/Titanic_in_Southampton.jpg/1920px-Titanic_in_Southampton.jpg)

The graph below is another histogram showing the distribution of the number of passengers and their age, but now in [Seaborn](https://seaborn.pydata.org/index.html), also a Python data visualization library.

```python
import seaborn as sns
```

```python
# Loading the dataset in seaborn data repository of Titanic
df = sns.load_dataset('titanic')

df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', kde=True, hue='sex')
plt.title('Age Distribution by Gender')
plt.show()
```

# 3. Graph using Plotly

![The sinking of the Titanic as depicted in Untergang der Titanic, a 1912 illustration by Willy Stwer.](https://upload.wikimedia.org/wikipedia/commons/thumb/6/6e/St%C3%B6wer_Titanic.jpg/300px-St%C3%B6wer_Titanic.jpg)

The graphs below show the number of passengers that survived and died but are now in [Plotly Express](https://plotly.com/python/plotly-express/), also a Python data visualization library.

```python
# Loading the dataset in seaborn data repository of Titanic and saving it in the 
titanic_data = sns.load_dataset('titanic')

# Viewing the first 5 rows
titanic_data.head()
```

```python
import plotly.express as px
```

```python
fig = px.pie(titanic_data, names='survived', title='Passenger Survival',color_discrete_map={'Not Survived': 'red', 'Survived': 'green'},labels={'SurvivalLabel': 'Survival'})
fig.show()
```

```python
fig = px.scatter(titanic_data, x='fare', y='age', color='survived', size='fare')
fig.show()
```
